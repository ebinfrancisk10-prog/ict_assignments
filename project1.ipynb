9. Text-Based Decision Game


```python
score = 0

print("Welcome to the Adventure Game!")

print("\nYou are at a crossroads.")
choice1 = input("Go left or right? ").lower()

if choice1 == "left":
    score += 10
    print("You find a treasure chest.")
else:
    score += 5
    print("You meet a traveler.")

print("\nYou see a cave.")
choice2 = input("Enter or skip? ").lower()

if choice2 == "enter":
    score += 10
    print("You find gold inside the cave!")
else:
    score -= 5
    print("You walk away safely.")

print("\nGame Over")
print("Your score:", score)

if score >= 20:
    print("Ending: You are a hero!")
else:
    print("Ending: You survived, but missed opportunities.")

```

3. Campus Store Inventory and Billing System


```python

inventory = {
    "Pen": {"price": 10, "stock": 20},
    "Notebook": {"price": 50, "stock": 10}
}

TAX = 0.05

def discount(subtotal):
    if subtotal >= 1000:
        return 0.10
    elif subtotal >= 500:
        return 0.05
    return 0

cart = []

print("\n--- Available Products ---")
for item in inventory:
    print(item, inventory[item])

while True:
    item = input("\nEnter product (or done): ")
    if item == "done":
        break
    if item not in inventory:
        print("Not available")
        continue

    qty = int(input("Quantity: "))
    if qty > inventory[item]["stock"]:
        print("Insufficient stock")
        continue

    cart.append((item, qty))
    inventory[item]["stock"] -= qty


subtotal = 0
print("\n--- BILL ---")
for item, qty in cart:
    cost = inventory[item]["price"] * qty
    subtotal += cost
    print(f"{item}: {qty} x {inventory[item]['price']} = {cost}")

disc = subtotal * discount(subtotal)
tax = subtotal * TAX
total = subtotal + tax - disc

print("\nSubtotal:", subtotal)
print("Discount:", disc)
print("Tax:", tax)
print("Total Payable:", total)


print("\n--- Low Stock Alert ---")
for item in inventory:
    if inventory[item]["stock"] <= 5:
        print(item, "is low on stock")

```
