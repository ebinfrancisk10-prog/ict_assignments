```python
for number in range(1, 101):
    if number % 3 == 0 and number % 5 == 0:
        print("FizzBuzz")
    elif number % 3 == 0:
        print("Fizz")
    elif number % 5 == 0:
        print("Buzz")
    else:
        print(number)
```

```python
def calculate_bill(units):
    total_bill = 0
    if units <= 100:
        total_bill = units * 5
    elif units <= 200:
        total_bill = (100 * 5) + ((units - 100) * 7)
    else:
        total_bill = (100 * 5) + (100 * 7) + ((units - 200) * 10)
    return total_bill

units_consumed_str = input("Enter units consumed: ")

try:
    units_consumed = int(units_consumed_str)
    if units_consumed < 0:
        print("Units consumed cannot be negative.")
    else:
        bill_amount = calculate_bill(units_consumed)
        print(f"Total Bill Amount: ₹{bill_amount}")
except ValueError:
    print("Invalid input. Please enter a whole number for units consumed.")
```

```python
def is_palindrome(text):
    # Convert to string to handle numbers correctly
    text_str = str(text)
    # Normalize the string: remove spaces and convert to lowercase
    cleaned_text = ''.join(char.lower() for char in text_str if char.isalnum())
    # Compare the cleaned string with its reverse
    return cleaned_text == cleaned_text[::-1]

# Get input from the user
user_input = input("Enter num or string to be checked: ")

# Check if it's a palindrome and print the result
if is_palindrome(user_input):
    print("Given num/string is Palindrome")
else:
    print("Given num/string is not Palindrome")
```
